```python

# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```


## 1. Load Dataset

Upload `Global_Pollution_Analysis.csv` in the same folder as this notebook.


```python

df = pd.read_csv("Global_Pollution_Analysis.csv")

df.head()
```

```python

# Dataset information

df.info()

print(df.shape)
```

```python

# Missing values check

df.isnull().sum()
```

```python

# Data cleaning

df = df.dropna()

df.head()
```


# 2. Feature Engineering

Create pollution categories and energy efficiency related features.


```python

# Pollution severity categories

df['Pollution_Level'] = pd.cut(
    df['AQI'],
    bins=[0,50,100,200,500],
    labels=['Low','Medium','High','Very High']
)

df.head()
```

```python

# Energy per capita feature

df['Energy_per_Capita'] = (
    df['Energy_Consumption'] /
    df['Population']
)

df.head()
```


# 3. Convert Dataset into Transactions

Apriori requires transaction-style data.


```python

transaction_columns = [
    'Country',
    'Pollution_Level'
]

transactions = df[transaction_columns].astype(str)

transactions = transactions.values.tolist()

transactions[:5]
```

```python

encoder = TransactionEncoder()

encoded = encoder.fit(transactions).transform(transactions)

apriori_df = pd.DataFrame(
    encoded,
    columns=encoder.columns_
)

apriori_df.head()
```


# 4. Apriori Frequent Itemset Extraction


```python

frequent_itemsets = apriori(
    apriori_df,
    min_support=0.1,
    use_colnames=True
)

frequent_itemsets
```


# 5. Association Rule Mining

Rules are generated using confidence threshold.


```python

rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.5
)

rules.head()
```

```python

# Important evaluation metrics

rules[
[
'antecedents',
'consequents',
'support',
'confidence',
'lift'
]
]
```


# 6. Visualizations


```python

plt.figure(figsize=(10,5))

plt.bar(
    frequent_itemsets['itemsets'].astype(str),
    frequent_itemsets['support']
)

plt.xticks(rotation=90)

plt.xlabel("Frequent Itemsets")
plt.ylabel("Support")

plt.title("Frequent Pollution Itemsets")

plt.show()
```

```python

plt.figure(figsize=(8,5))

plt.scatter(
    rules['support'],
    rules['confidence']
)

plt.xlabel("Support")
plt.ylabel("Confidence")

plt.title("Support vs Confidence")

plt.show()
```
